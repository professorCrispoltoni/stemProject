# 🤖 K-Nearest Neighbors (KNN)
### Un algoritmo di Machine Learning spiegato passo per passo

---

**KNN** è uno degli algoritmi di classificazione più semplici e intuitivi del Machine Learning.

> **Idea di base:** per classificare un nuovo punto, guarda i *K* punti più vicini nel dataset e assegna la classe più frequente tra di loro.

---

## 📦 Passo 1 — Importare le librerie

> Usiamo due librerie fondamentali:
> - **NumPy** → calcoli matematici (es. radice quadrata)
> - **Matplotlib** → grafici

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 📊 Passo 2 — I dati di addestramento

> Il nostro dataset ha **6 punti** con coordinate (x, y), ognuno appartenente a una **classe**:
> - **Classe 0** (🔴 rossa) → punti in basso a sinistra
> - **Classe 1** (🔵 blu) → punti in alto a destra

```python
# Coordinate dei punti (x, y)
punti = [
    [1, 2],
    [2, 3],
    [3, 3],
    [6, 5],
    [7, 7],
    [8, 6]
]

# Classe di appartenenza di ogni punto
classi = [0, 0, 0, 1, 1, 1]

print("Punti nel dataset:")
for i in range(len(punti)):
    print(f"  Punto {i+1}: {punti[i]}  →  Classe {classi[i]}")
```

## 📈 Passo 3 — Visualizzare i dati

> Disegniamo i punti sul piano cartesiano.
> Ogni colore rappresenta una classe diversa.

```python
plt.figure(figsize=(7, 5))

for i in range(len(punti)):
    x = punti[i][0]
    y = punti[i][1]

    if classi[i] == 0:
        colore = "red"
    else:
        colore = "blue"

    plt.scatter(x, y, color=colore, s=100, zorder=3)

# Aggiungiamo la legenda manuale
plt.scatter([], [], color="red",  label="Classe 0")
plt.scatter([], [], color="blue", label="Classe 1")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Dataset di addestramento")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

## ❓ Passo 4 — Il nuovo punto da classificare

> Abbiamo un **nuovo punto** (⬛ nero) di cui **non conosciamo la classe**.
>
> **Obiettivo:** usare KNN per capire se appartiene alla Classe 0 o alla Classe 1.

```python
nuovo_punto = [5, 4]

plt.figure(figsize=(7, 5))

for i in range(len(punti)):
    x = punti[i][0]
    y = punti[i][1]
    colore = "red" if classi[i] == 0 else "blue"
    plt.scatter(x, y, color=colore, s=100, zorder=3)

# Nuovo punto
plt.scatter(
    nuovo_punto[0],
    nuovo_punto[1],
    color="black",
    marker="X",
    s=200,
    zorder=4,
    label="Nuovo punto (?)"
)

plt.scatter([], [], color="red",  label="Classe 0")
plt.scatter([], [], color="blue", label="Classe 1")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Il nuovo punto da classificare")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

## 📐 Passo 5 — La distanza euclidea

> Per trovare i "vicini più vicini" dobbiamo misurare la **distanza** tra il nuovo punto e ogni altro punto del dataset.
>
> Usiamo la **distanza euclidea** (il classico teorema di Pitagora!):
>
> $$d = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}$$
>
> Più la distanza è piccola, più i due punti sono vicini.

```python
# Esempio: distanza tra il nuovo punto e il primo punto del dataset
x1, y1 = punti[0]
x2, y2 = nuovo_punto

distanza_esempio = np.sqrt((x1 - x2)**2 + (y1 - y2)**2)
print(f"Distanza tra {punti[0]} e {nuovo_punto}: {distanza_esempio:.2f}")
```

## 🔢 Passo 6 — Calcolare tutte le distanze

> Calcoliamo la distanza tra il **nuovo punto** e **tutti i punti** del dataset.
> Salviamo ogni distanza insieme alla classe corrispondente.

```python
distanze = []

for i in range(len(punti)):
    x1 = punti[i][0]
    y1 = punti[i][1]

    x2 = nuovo_punto[0]
    y2 = nuovo_punto[1]

    distanza = np.sqrt((x1 - x2)**2 + (y1 - y2)**2)
    distanze.append([distanza, classi[i]])

print("Distanze calcolate (non ordinate):")
for i, d in enumerate(distanze):
    print(f"  Punto {punti[i]} → distanza: {d[0]:.2f}  |  classe: {d[1]}")
```

## 🔃 Passo 7 — Ordinare le distanze

> Ordiniamo le distanze **dal più vicino al più lontano**.
> Così possiamo facilmente scegliere i *K* punti più vicini.

```python
distanze.sort()

print("Distanze ordinate (dal più vicino al più lontano):")
for i, d in enumerate(distanze):
    print(f"  {i+1}° più vicino  →  distanza: {d[0]:.2f}  |  classe: {d[1]}")
```

## 🎯 Passo 8 — Scegliere i K vicini

> Scegliamo il parametro **K** = quanti vicini considerare.
>
> 💡 **K è il parametro chiave di KNN!**
> - K piccolo → più sensibile ai singoli punti (rischio di overfitting)
> - K grande → classificazione più "stabile" ma meno precisa localmente
>
> Proviamo con **K = 3**.

```python
k = 3

vicini = distanze[:k]

print(f"I {k} vicini più vicini al punto {nuovo_punto}:")
for i, v in enumerate(vicini):
    print(f"  {i+1}° vicino  →  distanza: {v[0]:.2f}  |  classe: {v[1]}")
```

## 🗳️ Passo 9 — La votazione

> Ogni vicino "vota" per la propria classe.
> **Vince la classe con più voti!**
>
> Questo meccanismo si chiama **majority voting**.

```python
conta0 = 0
conta1 = 0

for v in vicini:
    classe = v[1]
    if classe == 0:
        conta0 += 1
    else:
        conta1 += 1

print(f"Voti per Classe 0: {conta0}")
print(f"Voti per Classe 1: {conta1}")

# Classe finale
if conta0 > conta1:
    classe_finale = 0
else:
    classe_finale = 1

print(f"\n✅ Classe assegnata al nuovo punto: {classe_finale}")
```

## 🖼️ Passo 10 — Visualizzazione finale

> Disegniamo il risultato: il nuovo punto classificato, i K vicini e le linee di distanza.

```python
# Ricostruiamo le distanze con le coordinate (per disegnarle)
distanze_con_coord = []
for i in range(len(punti)):
    x1, y1 = punti[i]
    x2, y2 = nuovo_punto
    distanza = np.sqrt((x1 - x2)**2 + (y1 - y2)**2)
    distanze_con_coord.append([distanza, classi[i], punti[i]])

distanze_con_coord.sort()
vicini_coord = distanze_con_coord[:k]

plt.figure(figsize=(7, 5))

# Linee verso i K vicini
for v in vicini_coord:
    px, py = v[2]
    plt.plot(
        [nuovo_punto[0], px],
        [nuovo_punto[1], py],
        color="gray", linestyle="--", linewidth=1.2, alpha=0.6
    )

# Tutti i punti
for i in range(len(punti)):
    colore = "red" if classi[i] == 0 else "blue"
    plt.scatter(punti[i][0], punti[i][1], color=colore, s=100, zorder=3)

# Cerchio attorno ai K vicini
for v in vicini_coord:
    px, py = v[2]
    colore = "red" if v[1] == 0 else "blue"
    plt.scatter(px, py, color=colore, s=200, edgecolors="black", linewidths=2, zorder=4)

# Nuovo punto con il colore della classe assegnata
colore_finale = "red" if classe_finale == 0 else "blue"
plt.scatter(
    nuovo_punto[0], nuovo_punto[1],
    color=colore_finale, marker="X", s=250,
    edgecolors="black", linewidths=1.5,
    zorder=5, label=f"Nuovo punto → Classe {classe_finale}"
)

plt.scatter([], [], color="red",  label="Classe 0")
plt.scatter([], [], color="blue", label="Classe 1")

plt.xlabel("x")
plt.ylabel("y")
plt.title(f"KNN con K={k} — Risultato: Classe {classe_finale}")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

print(f"Il nuovo punto {nuovo_punto} è stato classificato come: Classe {classe_finale}")
```

## 📝 Riepilogo — Come funziona KNN

| Passo | Cosa fa |
|-------|---------|
| 1️⃣ | Raccoglie i dati con le classi note |
| 2️⃣ | Riceve un nuovo punto da classificare |
| 3️⃣ | Calcola la distanza euclidea da tutti i punti |
| 4️⃣ | Ordina i punti per distanza crescente |
| 5️⃣ | Prende i primi **K** punti (i più vicini) |
| 6️⃣ | Conta i voti per ogni classe |
| 7️⃣ | Assegna la classe con più voti ✅ |

### ⚠️ Limitazioni di KNN
- Lento con dataset grandi (deve calcolare tutte le distanze ogni volta)
- Sensibile alla scala delle variabili (meglio normalizzare i dati)
- K va scelto con cura (di solito con **cross-validation**)
